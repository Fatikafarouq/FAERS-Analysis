```python
import pandas as pd

drug = pd.read_csv("DRUG25Q2.txt", delimiter = "$", encoding = "latin-1", low_memory = False)
outcome = pd.read_csv("OUTC25Q2.txt", delimiter = "$", encoding = "latin-1", low_memory = False)
death = outcome[outcome['outc_cod']== "DE"]

death_reports = pd.merge(drug, death, on = "primaryid")

death_count = death_reports['drugname'].value_counts().head(5)

print(death_count)
```
